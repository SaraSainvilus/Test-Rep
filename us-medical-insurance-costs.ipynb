```python
# U.S. Medical Insurance Costs
import csv
```

```python
#organizing rows into lists
with open("insurance.csv", newline = '') as ins_data:
    reader = csv.DictReader(ins_data)
    ages = []
    sex = []
    bmis = []
    children  = []
    smokers = []
    regions = []
    charges = []
    for row in reader:
        ages.append(row['age'])
        sex.append(row['sex'])
        bmis.append(row['bmi'])
        children.append(row['children'])
        smokers.append(row['smoker'])
        regions.append(row['region'])
        charges.append(float(row['charges']))


#organizing by cost        
ins_data = {}
index = range(len(ages))
for i in index:
    key = charges[i]
    value = {'age': ages[i], 'sex': sex[i], 'bmi': bmis[i], 'children': children[i], 'smoker': smokers[i], 'region': regions[i], 'cost': charges[i]}
    ins_data[key] =  value
print(ins_data)   
    
        
```

```python
#creating organization function
def org_func(data, var_name):
    org_func_dict = {}
    for item in data.values():
        variable = item[var_name]
        if variable not in org_func_dict:
            org_func_dict[variable] = []
        org_func_dict[variable].append(item)
    return org_func_dict
```

```python
#organizing by sex
sex_data = org_func(ins_data, 'sex')
#organizing by age
age_data = org_func(ins_data, 'age')
#organizing by children
child_data = org_func(ins_data, 'children')
#organizing by region
region_data = org_func(ins_data, 'region')
#organizing by smoker
smoke_data = org_func(ins_data, 'smoker')
#organizing by bmi
bmi_data = org_func(ins_data, 'bmi')
```

```python
#creating count function
def count_variable(variable_list):
    var_count = {}
    for item in variable_list:
        if item not in var_count:
            var_count[item] = 1
        else:
            var_count[item] += 1
    return var_count
```

```python
#count coverage by sex
sex_count = count_variable(sex)
print(sex_count)
#count coverage by smoker
smoke_count = count_variable(smokers)
print(smoke_count)
#count coverage by region
region_count = count_variable(regions)
print(region_count)
#count coverage by parents
parent_count = count_variable(children)
print(parent_count)
#count coverage by age
age_count_data = count_variable(ages)
print(age_count_data)
```

```python
#what percentage have children?

par_count = 0  
for key, value in parent_count.items():
    if key == '0':
        continue  
    else:
        par_count += value

total_insured = len(ages)

parent_percentage = round((par_count/total_insured)*100, 2)

print(str(par_count) + ' out of ' + str(total_insured) +' or ' + str(parent_percentage) + 
      '% of insured clients have children.')
```

```python
#what percentage of each sex smoke?
female_smoker = 0 
male_smoker = 0
for key, value_list in sex_data.items():
    for value in value_list:
        if key == 'female' and value['smoker'] == 'yes':
            female_smoker += 1
        elif key == 'male' and value['smoker'] == 'yes':
            male_smoker += 1

total_female_count = sex_count['female']
total_male_count = sex_count['male']

print(str(female_smoker) + ' out of ' + str(total_female_count) + ' total insured women smoke.')
print(str(male_smoker) + ' out of ' + str(total_male_count) + ' total insured men smoke.')

fem_smoke_pct = round((female_smoker/total_female_count)*100, 1)
male_smoke_pct = round((male_smoker/total_male_count)*100, 1)
        
print('The percentage of insured women that smoke is ' + str(fem_smoke_pct) + '%.')    
print('The percentage of insured men that smoke is ' + str(male_smoke_pct) + '%.') 

if male_smoke_pct > fem_smoke_pct:
    print('A greater percentage of insured men smoke compared to women.')
else:
    print('A greater percentage of insured women smoke compared to men.')
    
```

```python
#how many people are insured senior citizens?

#assuming senior citizens are 60+ years old.
ages.sort(reverse=True)
minimum_age = str(60)
senior_citizens = []

for age in ages:
    if age == minimum_age or age > minimum_age:
        senior_citizens.append(age)
print(senior_citizens)

sen_cit_amt = len(senior_citizens)

print("There are " + str(sen_cit_amt) + " insured senior citizens.")
```

```python
#On average, do senior citizens pay more or less compared to other age groups?
#first, let's build the age groups
adult_min_age = str(18)
adult_max_age = str(59)
senior_min_age = str(60)

#Calculate the average price for adults
adult_count = 0
adult_cost = 0
for key, value_list in age_data.items():
    if key >= adult_min_age and key < adult_max_age:
        for value in value_list:
            adult_cost += round(value['cost'])
            adult_count += 1

avg_adult_cost = round(adult_cost/adult_count)
print("The average insurance cost for adults 18 and older is $" + str(avg_adult_cost) + ".")

#Calculate the average price for seniors
senior_cost = 0

for key, value_list in age_data.items():
    if key >= senior_min_age:
        for value in value_list:
            senior_cost += round(value['cost'])

avg_senior_cost = round(senior_cost/sen_cit_amt)
print("The average insurance cost for seniors 60 and older is $" + str(avg_senior_cost) + ".")

difference = avg_senior_cost - avg_adult_cost
        
if avg_senior_cost > avg_adult_cost:
    print("Senior citizens 60 and older pay $" + str(difference) + 
          " more in insurance costs on average compared to adults 18 and older.")
else:
    print("Adults 18 and older pay $" + str(difference) + 
          " more in insurance costs on average compared to senior citizens 60 and older.")
```

```python
#On average, do smokers pay more in insurance costs compared to non-smokers?

#Calculate total cost for smokers and non-smokers
smoke_cost = 0
smoke_count = female_smoker + male_smoker
non_smoke_cost = 0
non_smoke_count = total_insured - smoke_count
for key, value_list in smoke_data.items():
    if key == 'yes':
        for value in value_list:
            smoke_cost += value['cost']
    else:
        for value in value_list:
            non_smoke_cost += value['cost']
print(smoke_cost)
print(non_smoke_cost)

#Average calculations
avg_smoke_cost = round(smoke_cost/smoke_count)
avg_non_smoke_cost = round(non_smoke_cost/non_smoke_count)
diff_smoke = avg_smoke_cost - avg_non_smoke_cost

if avg_smoke_cost > avg_non_smoke_cost:
    print("On average, smokers pay $" + str(diff_smoke) + 
          " more in insurance costs compared to non-smokers.")
else:
    print("On average, non-smokers pay $" + str(diff_smoke) + 
          " more in insurance costs compared to smokers.")
```

```python
#On average, does parents spend more in insurance or non-parents?

non_parent_count = total_insured - par_count
non_parent_cost = 0 
parent_cost = 0
for key, value_list in child_data.items():
    if key == '0':
        for value in value_list:
            non_parent_cost += value['cost']
    else:
        for value in value_list:
            parent_cost += value['cost']

avg_par_cost = round(parent_cost/par_count) 
avg_non_par_cost = round(non_parent_cost/non_parent_count)
diff_par = avg_par_cost - avg_non_par_cost

if avg_par_cost > avg_non_par_cost:
    print("On average, insured parents spend $" + str(diff_par) + 
          " more on insurance compared to insureds without children.")
else:
    print("On average, insureds without children spend $" + str(diff_par) + 
          " more on insurance compared to insured parents.")
```

```python
#On average, which BMI group spend more in insurance?

#Building BMI groups
#  underweight = 18.5 and under
#  healthy = 18.5 - 24.9
#  overweight = 25 - 29.9
#  obesity = 30 - 39.9
#  severe obesity = 40 and over

under_cost = 0
under_count = 0
healthy_cost = 0
healthy_count = 0
over_cost = 0
over_count = 0
obese_cost = 0
obese_count = 0
severe_cost = 0
severe_count = 0

for key, value_list in bmi_data.items():
    if key <= '18.5':
        for value in value_list:
            under_cost += value['cost']
            under_count += 1
        
    elif key > '18.5' and key <= '24.9':
        for value in value_list:
            healthy_cost += value['cost']
            healthy_count += 1
            
    elif key > '24.9' and key <= '29.9':
        for value in value_list:
            over_cost += value['cost']
            over_count += 1
            
    elif key > '29.9' and key <= '39.9':
        for value in value_list:
            obese_cost += value['cost']
            obese_count += 1
            
    elif key > '39.9':
        for value in value_list:
            severe_cost += value['cost']
            severe_count += 1

Avg_under = round(under_cost/under_count)
Avg_healthy = round(healthy_cost/healthy_count)
Avg_over = round(over_cost/over_count)
Avg_obese = round(obese_cost/obese_count)
Avg_severe = round(severe_cost/severe_count)

bmi_cost_dict = {}
bmi_cost_dict['underweight'] = Avg_under
bmi_cost_dict['healthy weight'] = Avg_healthy
bmi_cost_dict['overweight'] = Avg_over
bmi_cost_dict['obese'] = Avg_obese
bmi_cost_dict['severely obese'] = Avg_severe

print(bmi_cost_dict)

high_avg_cost = 0
for key, value in bmi_cost_dict.items():
    if value > high_avg_cost:
        high_avg_cost = value
        bmi_group = key
        
print("On average, people who are considered " + bmi_group  + 
      " pay more in insurance costs compared to other bmi groups.")        
```
